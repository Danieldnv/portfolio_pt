```python
MG=yf.download("MGLU3.SA", start='2019-01-01')
```

```python
plt.figure(figsize=(16,8))
plt.plot(MG['Close'], label='Ações Magalu', color='DarkBlue')
plt.plot(MG['Close'].rolling(15).mean(), label='Média móvel Magalu', color='red')
plt.title('Preços de fechamento')
plt.xlabel('Anos', fontsize=16)
plt.ylabel('Preços', fontsize=16)
plt.legend()
plt.show()
```
